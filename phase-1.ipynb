```python
import numpy as np
```

### creating array from list

```python
arr_1d = np.array([1,2,3,4,5])
print("1d array is : ", arr_1d)

arr_2d = np.array([[1,2,3], [4,5,6]])
print("2d array is : ", arr_2d)
```

### Creating Array from Scratch

```python
zeros = np.zeros((3,4))
print("zeros matrix : \n", zeros)

ones = np.ones((2,3))
print("ones matrix : \n", ones);

full = np.full((2,2), 6)
print("full matrix : \n", full);

random = np.random.random((4,5))
print("random matrix : \n", random);

sequence = np.arange(0, 10)
print("sequence matrix : \n", sequence);
```

### Vector, Matrix and Tensor 

```python
vector = np.array([1,2,3]);
print("vector: ", vector);

matrix = np.array([[1,2,3],
                   [4,5,6],])

print("matrix: ", matrix);

tensor = np.array([[[1,2], [3,4]],
                   [[5,6], [7,8]]])
print("tensor: ", tensor);
```
